```python
import tensorflow as tf
import pandas as pd
```

```python
base_dados = pd.read_csv ('/content/drive/MyDrive/UNIFAA/wine-dataset.csv', encoding = 'utf-8')
```

```python
base_dados = base_dados.drop(base_dados.index[1])
base_dados
```

```python
dados = base_dados.drop(columns=['Wine'])
alvo = base_dados['Wine']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
alvo = le.fit_transform(alvo)
alvo
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dados, alvo, test_size=0.20, random_state=42)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dense(64, activation=tf.nn.relu),
    tf.keras.layers.Dense(4, activation=tf.nn.softmax)
])

model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['Accuracy'])

model.fit(x_train, y_train, epochs=50)

```

```python
model.evaluate(x_test, y_test)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix

previsoes_prob = model.predict(x_test)

previsoes = np.argmax(previsoes_prob, axis=1)

matrix = confusion_matrix(y_test, previsoes)

print(matrix)
```
